# Perform Recommendation 

### Import Datasets

```python
import pandas as pd
df = pd.read_csv('cloths-rating.csv')
df.head(3)
```

### Clean Data

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['UserID'] = le.fit_transform(df['UserID'])
df.head()
```

### Pivot table of ProductId, UserID and New_score

```python
df_pivot = df.pivot_table(index='ProductID', columns='UserID', values='Rating').fillna(0)
df_pivot
```

### Sparse Matrix

```python
from scipy.sparse import csr_matrix
pivot_csr_matrix = csr_matrix(df_pivot.values)
print(pivot_csr_matrix)
```

### Fitting Data into NearestNeighbors Model

```python
from sklearn.neighbors import NearestNeighbors 

model_nearest = NearestNeighbors(metric='cosine', n_neighbors=20, radius=1)
model_nearest.fit(pivot_csr_matrix)
```

### Matrix of cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(df_pivot)
similarity_matrix
```

```python
product_ID = int(input("Enter ProductID as shown in datasets: "))
data = list(df_pivot.index)
print(data)
```

```python
query_index = data.index(product_ID)
print(query_index)
```

```python
similarity, indices = model_nearest.kneighbors(df_pivot.iloc[query_index,:].values.reshape(1,-1),n_neighbors=8)
print(similarity)
print(indices)
```

### Create dictionary and store recommendations of productID

```python
data_dict = {}
for i in range(0, len(similarity.flatten())):
    
    if i == 0:
        print('Recommendation for {0}:\n'.format(df_pivot.index[query_index]))
    else:
        data_dict[str(df_pivot.index[indices.flatten()[i]])] = float(similarity.flatten()[i])
        print(f'{df_pivot.index[indices.flatten()[i]]}, is similarity distance = with {similarity.flatten()[i]:.20f}:')
        
print(data_dict)
```
